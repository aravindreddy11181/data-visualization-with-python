```python
disease=pd.read_csv('training disease.csv')
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,y_train,y_test=train_test_split(disease.loc[:,'itching':'yellow_crust_ooze'],disease['prognosis'])
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
#y_pred=knn.predict(data_new)
#print("tst set prediction:\n {}".format(y_pred))
y_pred=knn.predict(X_test)
print("tst set prediction:\n {}".format(y_pred))

```

```python
print("test set score:{}".format(np.mean(y_pred==y_test)))
```

```python
print("test set score:{}".format(knn.score(X_test,y_test)))
```
